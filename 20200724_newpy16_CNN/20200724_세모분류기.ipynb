```python
train_datagen=ImageDataGenerator(rescale=1./255, 
                   rotation_range=15,#0~15도 사이의 각도로 회전
                   width_shift_range=0.1, #좌우로 각각 10픽셀 이내로 이동
                   height_shift_range=0.1, #상하로 각각 10픽셀 이내로 이동
                   shear_range=0.5, #변형(0.5 라디안 내외로 시계 반대방향 회전)
                   zoom_range=[0.8, 2.0], #최소 0.8배, 최대 2배 범위에서 확대
                   horizontal_flip=True,
                   vertical_flip=True
                  )
```

```python
img=load_img('손글씨데이터/train/triangle/triangle001.png')
x=img_to_array(img) # 24,24,3
x=x.reshape((1,)+x.shape) # 1은 이미지가 한개라는 의미
x.shape
```

```python
i=1
for batch in train_datagen.flow(x, batch_size=1,
                                save_to_dir='myimg',
                                save_prefix='tri',
                                save_format='png'):
    if i>50:
        break
    i+=1
```
