```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior() # 텐서플로우 버전 1로 낮추기 (버전 2 기능 사용불가하게 만들기)
```

```python
# multivariable linear regression
# 회귀식 : ydata = w1x1 + w2x2 + w3x3 + b
# 영수 : 

# 모의고사 점수
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
# 수능점수
y_data = [152., 185., 180., 196., 142.]

# 모델만들기 위한 변수작업 (그래프 정의)
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)

w1 = tf.Variable(tf.random_normal([1]))
w2 = tf.Variable(tf.random_normal([1]))
w3 = tf.Variable(tf.random_normal([1]))
b = tf.Variable(tf.random_normal([1]))

hf = x1*w1 + x2*w2 + x3*w3 +b

cost = tf.reduce_mean(tf.square(hf-y))
opt = tf.train.GradientDescentOptimizer(0.000001)
train = opt.minimize(cost)

# 그래프 실행
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(3001):
    _,cv,hfv = sess.run([train,cost,hf],
                feed_dict = {x1:x1_data , x2:x2_data , x3:x3_data ,y:y_data})
    if step%10==0:
        print(step, 'cost : ',cv,'\n prediction : ', hfv)

```

```python
print(sess.run(hf, feed_dict={x1:50, x2:50, x3:50}))
print(sess.run(hf, feed_dict={x1:90, x2:90, x3:90}))
print(sess.run(hf, feed_dict={x1:[50,90], x2:[50,90], x3:[50,90]}))
```

```python
# 행렬의 곱셈을 쓰면 수식이 강렬해진다.

# 리쉐이핑
x_data = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]
y_data = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]

# x는 플레이스 홀더로서 학습하고자 하는 데이터(실수)를 임시로 저장하기 위한 저장소 / shape은 [데이터수 가변 ,컬럼수 3]
x = tf.placeholder(tf.float32, [None,3])
y = tf.placeholder(tf.float32, [None,1])
w = tf.Variable(tf.random_normal([3,1]))
b = tf.Variable(tf.random_normal([1]))
```

```python
# hf = x*w + b 라고만 정의하고 싶다면 : 행렬곱 matmul 사용
hf = tf.matmul(x,w) + b
# 행렬 shape : (None, 3)(3,1) => (None,1)
# 결과 : hf = x1*w1 + x2*w2 + x3*w3 +b
```

```python
cost = tf.reduce_mean(tf.square(hf-y))
opt = tf.train.GradientDescentOptimizer(0.000001)
train = opt.minimize(cost)

# 그래프 실행
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(3001):
    _,cv,hfv = sess.run([train,cost,hf],
                feed_dict = {x:x_data ,y:y_data})
    if step%10==0:
        print(step, 'cost : ',cv,'\n prediction : ', hfv)
```

```python
print(sess.run(hf, feed_dict={x:[[90,90,90]]}))
```
